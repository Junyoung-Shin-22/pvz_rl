```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('./checkpoints/250526_141207/train.npz')

print(data.files)
score = data['score']
frames = data['iter']

assert score.shape == frames.shape
```

```python
x = np.arange(0, 100000, 1000)

fig, ax = plt.subplots()
ax1, ax2 = ax, ax.twinx()

fig.suptitle('Train Score of AC Agent')

ax1.set_xlabel('iterations')
ax1.set_ylabel('score')
ax1.plot(x, score, label='score', color='b')

ax2.set_ylabel('frames')
ax2.plot(x, frames, label='frames', color='g')

fig.legend()

print('max score:', score.max())
print('max frames:', frames.max())
```
